```python
Players_Raw = pd.read_csv(r'C:\Users\sarto\Downloads\england-premier-league-players-2018-to-2019-stats.csv')
Players_Raw
```

```python
Players_cleaned = Players_Raw.iloc[:,:47]
Players_cleaned
```

```python
DataToDrop = ['league', 'season', 'birthday_GMT','birthday']
Players_cleaned = Players_cleaned.drop(columns=[col for col in DataToDrop if col in Players_cleaned.columns])
Players_cleaned
```

```python
Players_cleaned = Players_cleaned.sort_values(by='Current Club')
Players_cleaned
```

```python
Teams = pd.read_csv(r"C:\Users\sarto\Downloads\england-premier-league-teams-2018-to-2019-stats.csv")
Teams
```

```python
Ranked_Teams = Teams.sort_values(by='wins', ascending=False)
Ranked_Teams
```

```python
Ranked_Teams.iloc[:,:70]
```

```python
todrop = ['team_name', 'country', 'season', 'matches_played','matches_played_home','matches_played_away','suspended_matches']
Ranked_Teams = Ranked_Teams.drop(columns=[col for col in todrop if col in Ranked_Teams.columns])
Ranked_Teams
```

```python
Ranked_Teams = Ranked_Teams.sort_values(by='common_name')
Ranked_Teams
```

```python
Teams = Ranked_Teams.rename(columns={"common_name": "Current Club"})
Teams
```

```python
Data = pd.merge(Players_cleaned, Teams, on ='Current Club', how='inner')
Data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = Data[['minutes_played_overall', 'minutes_played_home', 'minutes_played_away', 'appearances_overall', 'appearances_home', 'appearances_away', 
          'wins_home','wins_away', 'draws','draws_home','draws_away','losses','losses_home','losses_away' ,
          'points_per_game', 'points_per_game_home','points_per_game_away', 'league_position', 'league_position_home', 'league_position_away',    
            'performance_rank', 'goals_scored', 'goals_conceded', 'goal_difference', 'total_goal_count', 'total_goal_count_home', 'total_goal_count_away', 
            'goals_scored_home', 'goals_scored_away', 'goals_conceded_home', 'goals_conceded_away', 'goal_difference_home', 'goal_difference_away',  
            'minutes_per_goal_scored', 'minutes_per_goal_scored_home',  'minutes_per_goal_scored_away', 'minutes_per_goal_conceded',   
            'minutes_per_goal_conceded_home',  'minutes_per_goal_conceded_away',  'clean_sheets',  'clean_sheets_home_y', 'clean_sheets_away_y', 'btts_count', 
            'btts_count_home', 'btts_count_away', 'fts_count', 'fts_count_home',  'fts_count_away',  'first_team_to_score_count',   'first_team_to_score_count_home',
            'first_team_to_score_count_away',  'corners_total',   'corners_total_home',  'corners_total_away',  'cards_total', 'cards_total_home',    
            'cards_total_away', 'average_possession', 'average_possession_home', 'average_possession_away', 'shots', 'shots_home', 
            'shots_away',  'shots_on_target', 'shots_on_target_home', 'shots_on_target_away', 'shots_off_target', 'shots_off_target_home', 'shots_off_target_away']]

Y = Data['wins']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
predictions = model.predict(X_test)
```

```python
result = pd.DataFrame({'wins': Y_test, 'predicted_wins': predictions})
```

```python
sorted_teams = result.sort_values(by='predicted_wins', ascending=False)
```
